```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
cars = pd.read_csv("cars.csv")
cars
```

```python
X = cars[cars.columns[:-1]]
#X["cubicinches"] = cars['cubicinches'].astype(float)
```

```python
for i in X.columns:
    print(X[i].isnull().sum())
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3,
                   max_iter=300,
                   n_init=10,
                   init="k-means++",
                   random_state=0)
y_kmeans = kmeans.fit_predict(X.values)
```
